```python
res = dataset.groupby('Row_Date').sum()[prediction_features]

from bokeh.io import output_file, output_notebook, reset_output, show
from bokeh.layouts import gridplot
from bokeh.plotting import figure

reset_output()

figs = []

for col in res.columns:
    fig = figure(x_axis_type="datetime", background_fill_color="#fafafa", title=f"Feature {col} trend")
    fig.line(res.index, res[col], line_width=2, alpha=0.8, color="#53777a")
    figs.append(fig)

grid = gridplot(figs, ncols=5)

show(grid)
```

```python
sorted_dataset = dataset.sort_values('Row_Date')
gb = sorted_dataset[prediction_features].groupby(sorted_dataset.index)

amnt_max_changes_per_user = (gb.count()['FT_Current_Accounts'] - 1).clip(0)
amnt_positive_changes_per_user = gb.apply(lambda df: ((np.diff(df.values, axis=0) > 0).sum(axis=1) > 0).sum())

max_possible_score = amnt_positive_changes_per_user.sum() / amnt_max_changes_per_user.sum()
best_kaggle_score = 0.03140

print("Max possible map@k score on dataset:", max_possible_score)
print("Best Kaggle score:", best_kaggle_score)
print("Best possible out of max potential ratio:", best_kaggle_score / max_possible_score)
```

```python
def compute_binary_assymetric_correlation(feature_df):
    corr_mat = pd.DataFrame(columns=feature_df.columns, dtype='float64')

    for col1 in feature_df.columns:
        for col2 in feature_df.columns:
            f1, f2 = feature_df[col1], feature_df[col2]
            corr_mat.loc[col1, col2] = (f1 * f2).sum() / (f1.sum() or 1)
    
    return corr_mat

binary_corr = compute_binary_assymetric_correlation(pred_dataset)

sns.set(rc={'figure.figsize':(12,12)})
sns.heatmap(binary_corr, cmap='Blues').set_title('predictors assymetric binary correlation')
```

```python
def plot_categorical_feature_distribution(features, feature_name, height=500, width=1200):
    count_df = features.groupby(feature_name).count()
    hist = count_df[count_df.columns[0]].sort_values(ascending=False)

    fig = figure(plot_height=height, plot_width=width, x_range=list(hist.index.values), background_fill_color="#fafafa", title=f"{feature_name} histogram")
    fig.quad(bottom=0, top=hist, left=np.arange(0, len(hist))*1, right=np.arange(1, len(hist)+1)*1, 
             fill_color='blue', line_color='black')
    fig.left[0].formatter.use_scientific = False

    return fig
    
def plot_binary_feature_distribution(features, feature_name, height=300, width=400):
    count_df = features.groupby(feature_name).count()
    hist = count_df[count_df.columns[0]].sort_values(ascending=False)

    fig = figure(plot_height=height, plot_width=width, x_range=[-1,2], background_fill_color="#fafafa", title=f"{feature_name} histogram")
    fig.quad(bottom=0, top=hist, left=[-0.5,0.5], right=[0.5,1.5], 
             fill_color='blue', line_color='black')
    fig.left[0].formatter.use_scientific = False

    return fig

def visualize_predictor_diff(features, feature_name):
    print(f"Difference of predictor prevalency, {feature_name} feature")
    return dataset.groupby(feature_name)[prediction_features].mean().transpose()

def sort_by_symmetric_ratio(df, f1, f2):
    sorted_index = pd.Series(np.max(np.array([df[f1] / df[f2], df[f2] / df[f1]]), axis=0), index=df.index).sort_values(ascending=False).index
    return df.loc[sorted_index]
```
