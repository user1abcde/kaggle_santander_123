```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = pd.read_pickle('dataset/partial_train.pkl')
```

```python
dataset
```

```python
dataset.describe()
```

```python
print("amount of missing data in each column:")
dataset.isna().sum()
```

```python
prediction_features = [col for col in dataset.columns if 'FT_' in col]
pred_dataset = dataset[prediction_features]

print("10 most common predictors:")
(pred_dataset.sum() / len(pred_dataset)).sort_values(ascending=False)[:10]
```

```python
subset = pred_dataset[pred_dataset.sum(axis=1) > 0]
predictor_correlation = subset.corr()

sns.set(rc={'figure.figsize':(12,12)})
sns.heatmap(predictor_correlation).set_title('predictors correlation')
```

```python
subset = dataset[pred_dataset.sum(axis=1) > 0]
feature_correlation = subset.corr().loc[prediction_features][['Gross_Household_Income', 'Age', 'Customer_Seniority_Months']]

sns.set(rc={'figure.figsize':(12,12)})
sns.heatmap(feature_correlation).set_title('features correlation')
```

```python
res = dataset.groupby('Row_Date').sum()[prediction_features]

from bokeh.io import output_file, output_notebook, reset_output, show
from bokeh.layouts import gridplot
from bokeh.plotting import figure

reset_output()

figs = []

for col in res.columns:
    fig = figure(x_axis_type="datetime", background_fill_color="#fafafa", title=col)
    fig.line(res.index, res[col], line_width=2, alpha=0.8, color="#53777a")
    figs.append(fig)

grid = gridplot(figs, ncols=5)

show(grid)
```
